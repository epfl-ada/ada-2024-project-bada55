```python
# Coefficients pour les deux sites



df = pd.DataFrame({
    'topic': model_ba.params.index,
    'Beer Advocate': [val * 100 for val in model_ba.params.values],  # Convertir en %
    'Rate Beer': [val * 100 for val in model_rb.params.values]   # Convertir en %
})

fig = go.Figure()

fig.add_trace(go.Bar(
    x=df['topic'],
    y=df['Beer Advocate'],
    name='Beer Advocate',
    marker_color='lightcoral', 
    # marker_line_color='red',  
    marker_line_width=1.5,
    text=[f"{val:.1f}%" for val in df['Beer Advocate']]
))

fig.add_trace(go.Bar(
    x=df['topic'],
    y=df['Rate Beer'],
    name='Rate Beer',
    marker_color='lightblue',  
    # marker_line_color='blue',  
    marker_line_width=1.5,
    text=[f"{val:.1f}%" for val in df['Rate Beer']]
))

fig.update_layout(
    title='Comparison of coefficients by topic (in %)',
    xaxis_title='topic',
    yaxis_title='Coefficient (%)',
    barmode='group',  
    legend_title='Site',
    template='plotly_white',
    yaxis=dict(ticksuffix='%') 
)

fig.write_html("../generated/figures/rating_coeff.html")
```

```python
ba_coeff = pd.DataFrame({
    'Coeff': model_ba.params,
    'topic': ['appearance', 'aroma', 'palate', 'taste', 'overall'],
    'site': 'Beer Advocate'
})

rb_coeff = pd.DataFrame({
    'Coeff': model_rb.params,
    'Theme': ['appearance', 'aroma', 'palate', 'taste', 'overall'],
    'site': 'Rate Beer'
})

df_combined = pd.concat([ba_coeff, rb_coeff], axis=0).reset_index()
print(df_combined)
fig = px.bar(df_combined,
             x="site", 
             y="Coeff", 
             color="index",  
             text=df_combined["Coeff"].apply(lambda x: f"{x*100:.1f}%"),  
             labels={'Coeff': 'Pourcentage (%)'},  
             title="Comparison of coefficients by topic (in %)")

fig.update_layout(
    yaxis=dict(tickformat='.0%', range=[0, 1]),  
    barmode='stack',  
    xaxis_title='Site',
    yaxis_title='Pourcentage (%)',
    template='plotly_white',
)

fig.update_traces(texttemplate='%{text}', textposition='inside')

fig.write_html("../generated/figures/rating_coeff2.html")
```

```python
ba_params = np.array([0.5, 0.200, 0.100080, 0.400013, 0.199913])  # Paramètres pour Beer Advocate
rb_params = np.array([0.5, 0.2, 0.088889, 0.200000, 0.422223])  # Paramètres pour Rate Beer

# Erreurs standards des coefficients (si disponibles)
ba_se = model_ba.bse.values  # Erreurs standard pour Beer Advocate (exemple)
rb_se = model_rb.bse.values  # Erreurs standard pour Rate Beer (exemple)

# Calcul des t-statistics pour chaque thème
t_stats = (ba_params - rb_params) / np.sqrt(ba_se**2 + rb_se**2)

# Si tu veux tester si ces différences sont significatives, tu peux aussi calculer les p-values :
# Par exemple, en supposant une distribution t avec un certain nombre de degrés de liberté (n-1 si les coefficients sont indépendants)
n = len(ba_params)  # nombre de coefficients
p_values = [stats.t.sf(np.abs(t), df=n-1) * 2 for t in t_stats]  # p-value pour test bilatéral

# Afficher les résultats pour chaque thème
for i, (t_stat, p_value) in enumerate(zip(t_stats, p_values)):
    print(f"Theme {i+1}: t-statistic = {t_stat:.3f}, p-value = {p_value:.3f}")
```
